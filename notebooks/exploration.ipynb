```python
import os
import matplotlib.pyplot as plt
from PIL import Image

base_dir = "data"

for split in ["train", "val", "test"]:
    print(f"\n{split.upper()}:")
    for label in ["NORMAL", "PNEUMONIA"]:
        folder = os.path.join(base_dir, split, label)
        count = len(os.listdir(folder))
        print(f"{label}: {count} images")
```

```python
import random

def show_images(image_paths, title):
    plt.figure(figsize=(10, 5))
    for i, path in enumerate(image_paths):
        img = Image.open(path)
        plt.subplot(1, len(image_paths), i + 1)
        plt.imshow(img.convert('L'), cmap='gray')
        plt.title(title)
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Random examples
normal_paths = [os.path.join(base_dir, 'train/NORMAL', f) for f in os.listdir(base_dir + '/train/NORMAL')]
pneumonia_paths = [os.path.join(base_dir, 'train/PNEUMONIA', f) for f in os.listdir(base_dir + '/train/PNEUMONIA')]

show_images(random.sample(normal_paths, 3), "Normal")
show_images(random.sample(pneumonia_paths, 3), "Pneumonia")
```
